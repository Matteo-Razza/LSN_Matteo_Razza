Use the NSL simulator code to <span style="color:red">print out instantaneous values of potential energy per particle, $U/N$, along **equilibrated** Monte Carlo simulations</span> (of $M=5\times 10^5$ MC steps and where the Metropolis acceptance ratio has been fixed approximately to 50%) in the following thermodynamic conditions:

1. I computed instantaneous values of potential energy per particle, but the system needed to be equilibrated. Therefore, for the gas state, $200000$ initial steps are performed, while for the liquid and solid states, only $5000$ steps are required. In all cases, $5 \cdot 10^5$ Monte Carlo steps are then executed. We want the Metropolis acceptance ratio to be approximately $50\%$ to respect an empirical and reasonable $50\%$ rule. To achieve this, the Metropolis algorithm was implemented with $\delta_{\text{solid}} = 0.055$, $\delta_{\text{liquid}} = 0.100$, and $\delta_{\text{gas}} = 6.000$. The acceptance ratio for $\delta_{\text{gas}} = 6.000$ is $61\%$, but I couldn't find any better value.

    After having produced this data, I used them to make pictures of the autocorrelation functions of $U/N$ to approximately estimate how large the blocks in my simulations should be.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_autocorrelation(filename, t_values, tmax=500000):
    # Apro il file in modalità di lettura
    with open(filename, 'r') as input_file:
        # Skippo la riga del titolo
        input_file.readline()
        # Leggo le altre righe
        lines = input_file.readlines()

    # Inizializzo una lista per memorizzare i secondi valori di tutte le righe
    second_values = []

    # Itero attraverso ogni riga del file
    for line in lines:
        # Divido la riga in base agli spazi
        values = line.split()
        # Estraggo il secondo valore e lo aggiungo alla lista
        second_values.append(float(values[1]))

    sum_values = 0.0
    sum2_values = 0.0

    for i in range(tmax):
        sum_values += second_values[i]
        sum2_values += second_values[i] ** 2

    den = 1.0 / tmax * sum2_values - (1.0 / tmax * sum_values) ** 2

    ac_values = []

    for t in t_values:
        sum_1 = 0
        sum_2 = 0
        sum_3 = 0
        num = 0
        ac = 0

        for i in range(1, tmax - t):
            a = second_values[i]
            b = second_values[i+t]
            sum_1 += a * b
            sum_2 += a
            sum_3 += b

        num = 1.0 / (tmax - t) * sum_1 - 1.0 / (tmax - t) * sum_2 * 1.0 / (tmax - t) * sum_3
        ac = num / den
        ac_values.append(ac)

    return ac_values

# Parametri comuni
t_values = [1, 2, 4, 5, 8, 10, 20, 25, 40, 50, 80, 100, 125, 200]
tmax = 500000

# Calcola e plotta per ogni file
files = ['solid_autocorrelation.dat', 'liquid_autocorrelation.dat', 'gas_autocorrelation.dat']
labels = ['Solid', 'Liquid', 'Gas']

for filename, label in zip(files, labels):
    ac_values = calculate_autocorrelation(filename, t_values, tmax)
    plt.plot(t_values, ac_values, label=label)

plt.title('Estimation of the length of the blocks for different states')
plt.xlabel('Block length $L$')
plt.ylabel('Autocorrelation of $U/N$')
plt.legend()
plt.show()
```

The gas phase needs very few steps per block for them to be uncorrelated. The solid phase seems to need blocks of about $100$ steps, while the liquid phase has the longest autocorrelation, needing about $200$ steps per block. It is safe to say that with $2000$ steps per block, every measurement obtained is surely independent.

2. Here it is presented an estimation of the statistical uncertainties of the expectation value of $U/N$ for different size of the blocks in the data blocking technique, from $L=M/N=10$ to $L=5 \cdot 10^3$, being $N$ the number of blocks. I used the exact same generated data.

```python
def calculate_error(filename, L, tmax=500000):
    # Apro il file in modalità di lettura
    with open(filename, 'r') as input_file:
        # Skippo la riga del titolo
        input_file.readline()
        # Leggo le altre righe
        lines = input_file.readlines()

    # Inizializzo una lista per memorizzare i secondi valori di tutte le righe
    second_values = []

    # Itero attraverso ogni riga del file
    for line in lines:
        # Divido la riga in base agli spazi
        values = line.split()
        # Estraggo il secondo valore e lo aggiungo alla lista
        second_values.append(float(values[1]))

    error = []
    
    for l in L:
        N = int(tmax/l)
        average = [0] * N
        sum_average = 0
        sum_average_2 = 0
        
        for j in range(N):
            for i in range(l):
                average[j] += second_values[j*l+i] / float(l)

        for j in range(N):
            sum_average += average[j]
            sum_average_2 += average[j] * average[j]
            
        error.append(np.sqrt(np.abs(sum_average_2 / float(N) - (sum_average / float(N))**2) / float(N)))

    return error

# Parametri comuni
L = [i for i in range(10, 5001, 10)]
tmax = 500000

# File e label
files = ['solid_autocorrelation.dat', 'liquid_autocorrelation.dat', 'gas_autocorrelation.dat']
labels = ['solid', 'liquid', 'gas']

# Calcola gli errori per ogni file
errors = []
for filename in files:
    errors.append(calculate_error(filename, L, tmax))

# Crea la griglia di subplot
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Plot singoli per solid, liquid, gas
for i, (error, label) in enumerate(zip(errors, labels)):
    row, col = divmod(i, 2)
    axs[row, col].plot(L, error)
    axs[row, col].set_title(f'Estimation of the statistical uncertainties for {label}')
    axs[row, col].set_xlabel('Block length $L$')
    axs[row, col].set_ylabel('Error on $U/N$')

# Plot con tutti e tre insieme
for error, label in zip(errors, labels):
    axs[1, 1].plot(L, error, label=label)

axs[1, 1].set_title('Estimation of the statistical uncertainties for all states')
axs[1, 1].set_xlabel('Block length $L$')
axs[1, 1].set_ylabel('Error on $U/N$')
axs[1, 1].legend()

# Layout
plt.tight_layout()
plt.show()
```

As we can see, for block sizes smaller than the autocorrelation time the statistical uncertainty is smaller. Then it increases until autocorrelation saturates and the error converges to a mean value.

By means of your upgraded NSL simulator code, equilibrate and <span style="color:red">perform MC NVT simulations via a Lennard-Jones model</span> of Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) in the following conditions:
1. solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

show in pictures the obtained average values and uncertainties for the potential energy per particle, $U/N$, the pressure $P$ and the radial distribution function $g(r)$ in LJ units ... <span style="color:red"> and compare your MC results for the radial distribution function, $g(r)$, with those obtained with Molecular Dynamics NVE simulations in similar thermodynamic conditions.</span>

I performed Monte Carlo NVT simulations in the same conditions as previously illustated: 
- solid phase: equilibration time = $5000$ steps, $\delta_{\text{solid}}=0.055$
- liquid phase: equilibration time = $5000$ steps, $\delta_{\text{liquid}} = 0.100$
- gas phase: equilibration time = $200000$ steps, $\delta_{\text{gas}} = 6.000$

Here I show pictures of potential energy per particle $U/N$, pressure $P$ and radial distribution function $g(r)$ (in LJ units). For each one of these properties, Monte Carlo NVT results are compared with Molecular Dynamics NVE. Of course I performed the simulations in similar thermodynamic conditions (for NVE settings see Exercises 04). I always used $20$ blocks and $2000$ step per block.

```python
# Caricamento dei dati
x, y, z = np.loadtxt("solid_potential_energy_NVT.dat", usecols=(0, 2, 3), unpack=True)
a, b, c = np.loadtxt("solid_pressure_NVT.dat", usecols=(0, 2, 3), unpack=True)
i, j, k = np.loadtxt("solid_gofr_NVT.dat", usecols=(0, 1, 2), unpack=True)
o, p, q = np.loadtxt("solid_potential_energy_NVE.dat", usecols=(0, 2, 3), unpack=True)
r, s, t = np.loadtxt("solid_pressure_NVE.dat", usecols=(0, 2, 3), unpack=True)
l, m, n = np.loadtxt("solid_gofr_NVE.dat", usecols=(0, 1, 2), unpack=True)

# Creazione della griglia di subplot
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Titolo sopra l'intera griglia di subplot
fig.suptitle('MC NVE - MD NVT simulations via a Lennard-Jones model of Argon in solid phase')

# Grafico per l'energia potenziale
axs[0, 0].errorbar(x, y, yerr=z, label='NVT')
axs[0, 0].errorbar(o, p, yerr=q, label='NVE')
axs[0, 0].set_title('Potential energy per particle, $U/N$')
axs[0, 0].set_xlabel('Block')
axs[0, 0].set_ylabel('Potential energy')
axs[0, 0].set_xticks(np.arange(min(x), max(x), 2))  # Imposta ticks interi sull'asse x
axs[0, 0].legend()

# Grafico per la pressione
axs[0, 1].errorbar(a, b, yerr=c, label='NVT')
axs[0, 1].errorbar(r, s, yerr=t, label='NVE')
axs[0, 1].set_title('Pressure, $P$')
axs[0, 1].set_xlabel('Block')
axs[0, 1].set_ylabel('Pressure')
axs[0, 1].set_xticks(np.arange(min(x), max(x), 2))  # Imposta ticks interi sull'asse x
axs[0, 1].legend()

# Grafico per la funzione di distribuzione radiale NVT
axs[1, 0].errorbar(i, j, yerr=k)
axs[1, 0].set_title('Radial distribution function $g(r)$ (NVT)')
axs[1, 0].set_xlabel('r/' r'$\sigma$')
axs[1, 0].set_ylabel('Radial distribution function')

# Grafico per la funzione di distribuzione radiale NVE
axs[1, 1].errorbar(l, m, yerr=n)
axs[1, 1].set_title('Radial distribution function $g(r)$ (NVE)')
axs[1, 1].set_xlabel('r/' r'$\sigma$')
axs[1, 1].set_ylabel('Radial distribution function')

# Aggiungi la griglia a tutti i grafici
for ax in axs.flat:
    ax.grid(True)

# Ottimizzazione del layout
plt.tight_layout()

# Visualizzazione
plt.show()
```

It shouldn't be too surprising to see the difference between Monte Carlo NVT and Molecular Dynamics NVE simulations regarding the measurements of potential energy and pressure. As already mentioned, we are not simulating the systems under the exact same conditions: I chose specific values for the initial temperature in the NVE simulations based on the trends plotted (see Exercise 4), while in the NVT simulations, it is, by definition, fixed. Microcanical ensemble and Canonical ensemble coincide only in the true thermodinacal limit, i.e., when the number of particles tends to infinity. Even if we consider periodic boundary condition to simulate this limit, there are still some differences. What's really surprising is the great agreement in the radial distribution function. It's very remarkable how similar they are, given the intrinsic differences between the two approaches: they both effectively capture the spatial arrangement of particles within the system.
