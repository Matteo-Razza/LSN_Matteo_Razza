### Exercise 04.1

The NSL_SIMULATOR code we have provided compute the total (internal) energy per particle, $E/N$, the potential energy per particle, $U/N$, the kinetic energy per particle, $K/N$, and the temperature, $T$.

<span style="color:red">Include the calculation of the pressure, $P$, inside your Molecular Dynamics code: block averages, progressive average values and progressive uncertainties.</span>

Remember that, via the **virial theorem**, the pressure in a Lennard-Jones system can be estimated by means of:

$$ P = \rho k_B \langle T \rangle + \frac{1}{3V} \left\langle \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 48\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle $$

### Exercise 04.2

By means of your upgraded code, <span style="color:red">perform MD simulations via a Lennard-Jones model</span> of Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) in the following conditions:
1. solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

and show in pictures the obtained progressive average values and uncertainties as a functions of the increasing number of blocks for the total (internal) energy per particle, $E/N$, the potential energy per particle, $U/N$, the kinetic energy per particle, $K/N$, the temperature, $T$, and the pressure, $P$, in reduced units!

```python
import matplotlib.pyplot as plt
import numpy as np

# Caricamento dei dati
x, y, z = np.loadtxt("solid_total_energy.dat", usecols=(0, 2, 3), unpack=True)
a, b, c = np.loadtxt("solid_potential_energy.dat", usecols=(0, 2, 3), unpack=True)
i, j, k = np.loadtxt("solid_kinetic_energy.dat", usecols=(0, 2, 3), unpack=True)
l, m, n = np.loadtxt("solid_temperature.dat", usecols=(0, 2, 3), unpack=True)
o, p, q = np.loadtxt("solid_pressure.dat", usecols=(0, 2, 3), unpack=True)

# Creazione della griglia di subplot
fig, axs = plt.subplots(3, 2, figsize=(10, 10))

# Titolo sopra l'intera griglia di subplot
fig.suptitle('MD simulation via a Lennard-Jones model of Argon in solid phase (without equilibration)')

# Grafico per l'energia totale
axs[0, 0].errorbar(x, y, yerr=z)
axs[0, 0].set_xlabel('# of blocks')
axs[0, 0].set_ylabel('Total energy')
axs[0, 0].set_title('Total (internal) energy per particle, $E/N$')

# Nascondi il subplot in posizione (0, 1)
axs[0, 1].axis('off')

# Grafico per l'energia potenziale
axs[1, 0].errorbar(a, b, yerr=c)
axs[1, 0].set_title('Potential energy per particle, $U/N$')
axs[1, 0].set_xlabel('# of blocks')
axs[1, 0].set_ylabel('Potential energy')

# Grafico per l'energia cinetica
axs[1, 1].errorbar(i, j, yerr=k)
axs[1, 1].set_title('Kinetic energy per particle, $K/N$')
axs[1, 1].set_xlabel('# of blocks')
axs[1, 1].set_ylabel('Kinetic energy')

# Grafico per la temperatura
axs[2, 0].errorbar(l, m, yerr=n)
axs[2, 0].set_title('Temperature, $T$')
axs[2, 0].set_xlabel('# of blocks')
axs[2, 0].set_ylabel('Temperature')

# Grafico per la pressione
axs[2, 1].errorbar(o, p, yerr=q)
axs[2, 1].set_title('Pressure, $P$')
axs[2, 1].set_xlabel('# of blocks')
axs[2, 1].set_ylabel('Total energy')

# Aggiungi la griglia a tutti i grafici
for ax in axs.flat:
    ax.grid(True)

# Ottimizzazione del layout
plt.tight_layout()

# Visualizzazione
plt.show()
```

### <span style="color:green"> Equilibration is needed!</span>

By monitoring the temperature, $T$, during the first time steps of the simulation, you will observe that at the beginning, starting from a perfect crystalline configuration, $T$ will not conserve its value. Can you understand and explain this behaviour?

Explore and find the starting value of $T$ needed to stabilize, after a certain number of time steps (equilibration time), the average value of $T$ on the desired $T^\star$. <span style="color:red">Perform your solid/liquid/gas phase MD simulations by re-starting them after such equilibration time.</span>

In the following graph, the temperature of Argon in its gas state is plotted, starting from an initial temperature of $0.45$ in reduced units. The first $4000$ blocks, each containing $100$ steps, are displayed. It is evident that choosing such an initial temperature ensures a consistent temperature of $1.2$ in reduced units after about $2000$ blocks. This is why it makes more sense to perfomr molecular dynamics (MD) simulations for the gas phase only after $200000$ steps, allowing for equilibration. Following this equilibration period, the average value of $T$ stabilizes at the desired $T^\star$. 

```python
import matplotlib.pyplot as plt
import numpy as np

# Caricamento dei dati
x, y, z = np.loadtxt("temp_gas.dat", usecols=(0, 2, 3), unpack=True)

# Plotto i dati
plt.errorbar(x, y, yerr=z) 
plt.title('Temperature, $T$') 
plt.xlabel('# of blocks') 
plt.ylabel('Temperature') 
plt.grid(True) 

# Visualizzazione
plt.show()
```
