<span style="color:red">Include the calculation of the pressure, $P$, inside your Molecular Dynamics code: block averages, progressive average values and progressive uncertainties.</span>

Remember that, via the **virial theorem**, the pressure in a Lennard-Jones system can be estimated by means of:

$$ P = \rho k_B \langle T \rangle + \frac{1}{3V} \left\langle \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 48\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle $$

The first observation that catches our attention is that the temperature does not maintain the initially set value. This behavior occurs because, initially, we place the particles in a crystal configuration and allow the system to evolve while keeping the number of particles $N$ and the volume $V$ constant (thanks to periodic boundary conditions). More importantly, we maintain the same *total* energy $E$ (not just the kinetic one!), which is ensured by the Verlet algorithm. In other words we are working in a microcanonical ensemble which is a statistical ensemble that describes isolated systems, i.e., systems with a defined value of energy, volume, and number of particles.

This energy conservation is essential because we are interested in statistical properties rather than precise configurational trajectories. However, the trade-off is that we need to allow some time for the system to reach equilibrium, as explained in the next paragraph. Once reached equilibrium, the ergoicity of the microcanonical ensemble ensures that ensemble averages at equilibrium do not depend on time.

## Observations
Here I present some observations that arose while coding.
- The most critical part of the Verlet algorithm probably lies in the initial previous-position initialization. In fact, the Verlet algorithm needs two positions for each particle at every step: one at $t-\delta t$ (labeled "old") and the actual position $\vec{r}(t)$. However, initially we only have the starting positions and velocities, so we determine $\vec{r}(0-\delta t)$ with a Taylor expansion: $\vec{r}(0-\delta t)=\vec{r}(t)-\vec{v}(t) \delta t$. Given that this artifice is probably the main source of error, my idea was to improve it by going to a higher order expansion: $\vec{r}(0-\delta t)=\vec{r}(t)-\vec{v}(t) \delta t + \frac{1}{2} \delta t^2 \vec{F}(\vec{r}(0))$, where $\vec{F}(\vec{r}(0))$ is the force on the particle examined at $t=0$.
- Another observation moving towards this direction regards the possibility given by the code to eventually restart the simulation from an existing file. This is achievable by setting RESET in the input file to $1$. However, this alternative would again use the previously illustrated algorithm to determine $\vec{r}(0-\delta t)$. I thought about improving the reset option by storing not only the last positions but also the second last ones. In this way, when we restart from a final configuration obtained by a previous run of the code, there is no need to produce any other fictitious initial data.
- I found some problems in the measure function. The code gave the possibility to choose before running the code which properties would be computed. However, removing some properties from the respective input file was very dangerous and would ruin every measurement. After some debugging, I understood that the code was trying to compute even those properties that I had removed from the input file. So, I carefully examined every boolean _measurement_property label and added some checks to ensure that if a certain property was not in the input file, it wouldn't be computed. I spent a lot of time on this bug because the terminal didn't raise any segmentation fault errors.
- After the first measurement, I plotted some pictures and noticed that some properties (typically the total internal energy per particle) presented straight lines and step-like changes in the fluctuations. This was clearly a rounding error: the precision was not high enough to evaluate the small changes that actually occurred. So, I added some "setprecision" functions in the code to better appreciate what was going on.
- Furthermore, I tried to improve another aspect of the code regarding the Verlet algorithm. Positions are determined by a third-order Taylor expansion:
  $$\vec{r}(t + \delta t) \approx \vec{r}(t) + \delta t \vec{v}(t) + \frac{1}{2} \delta t^2 \vec{a}(t) + \frac{1}{6} \delta t^3 \dot{\vec{a}}(t)$$
  $$\vec{r}(t - \delta t) \approx \vec{r}(t) - \delta t \vec{v}(t) + \frac{1}{2} \delta t^2 \vec{a}(t) - \frac{1}{6} \delta t^3 \dot{\vec{a}}(t)$$
  $$\vec{r}(t + \delta t) \approx 2\vec{r}(t) - \vec{r}(t - \delta t) + \delta t^2 \vec{a}(t)$$
  So the new position at time $t + \delta t$ uses the positions at time $t$ and $t - \delta t$. In the provided code, we computed the new velocities as
  $$\vec{v}(t) \approx \frac{\vec{r}(t + \delta t) - \vec{r}(t - \delta t)}{2 \delta t}$$
  It is clear that there is a shift between positions and velocities in the code of a time interval $\delta t$. So my idea was to determine $\vec{v}(t + \delta t)$:
  $$\vec{v}\biggl(t + \frac{\delta t}{2}\biggr) \approx \frac{\vec{r}(t + \delta t) - \vec{r}(t)}{\delta t}$$
  $$\vec{v}(t + \delta t) \approx \vec{v}\biggl(t + \frac{\delta t}{2}\biggr) + \frac{\delta t}{2} \vec{F}\biggl(\vec{r}\biggl(t + \frac{\delta t}{2}\biggr)\biggr) \approx \vec{v}\biggl(t + \frac{\delta t}{2}\biggr) + \frac{\delta t}{2} \vec{F}(\vec{r}(t))$$

  Actually, this change didn't affect the results as much as I thought, but conceptually it is important to compute potential and kinetic energy at the same time, given that the Verlet algorithm works by preserving the total energy.

- As far as the total energy is concerned, it is reassuring that it seems to be preserved over time. In some simulations not reported here, I found that it keeps its original value proportionally to $\delta t^2$.

- At last, a little comment on the behavior of temperature is in order. It is perfectly normal that the system needs an equilibration time: at first, we just start from a crystalline configuration. As far as the solid and liquid simulations are concerned, it is clear that the initial configuration presents a lower potential energy than any other one. So in the first steps, we expect an increase in the potential energy and a consequent decrease in the kinetic energy, hence in the temperature. Conversely, the gas phase has too low a density: particles are too far from one another, so they minimize potential energy by increasing kinetic energy. This explains why we see a rise in temperature even if the gas phase remains really unstable.
